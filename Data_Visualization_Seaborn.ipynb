```python
## Data Visualization using Seaborn

import seaborn as sns

```

```python
## Basic ploting
tips=sns.load_dataset('tips')
tips
```

```python
##Scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs Tips ")

plt.show()
```

```python
##Line Plot
sns.lineplot(x='size',y='total_bill', data=tips)
plt.title("Line Plot of Total Bill by Size")
plt.show()
```

```python
## Categorical Plot the variables contain sex, smoker, day, time etc..
##Bar Plot- on the basis of day i want to show total bills
sns.barplot(x='day',y='total_bill', data=tips)
plt.title('Bar plot of Total Bill by Day')
plt.show()
```

```python
## Box plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Box Plot of Total Bill by day')
plt.show()
```

```python
## Violin Plot
sns.violinplot(x='day',y='total_bill', data=tips)
plt.title('Violin plot of Total Bill by Day')
plt.show()
```

```python
## Histogram
sns.histplot(tips['total_bill'],bins=10,kde=True)

```

```python
sns.histplot(tips['total_bill'],bins=10,kde=False)
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'], fill=True)
```

```python
#PairPlot
sns.pairplot(tips)
```

```python
##HeatMap
corr =tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
