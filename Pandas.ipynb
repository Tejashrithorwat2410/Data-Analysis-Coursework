```python
## Data aggregating and grouping

data=pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
grouped_mean = data.groupby('total_rooms')['population'].mean()
grouped_mean
```

```python
grouped_agg = data.groupby('total_rooms')['population'].agg(['mean','sum','max','min','count'])
grouped_agg
```

```python
## Merging and Joining Dataframe
df1 = pd.DataFrame({'key':[ 'A','B','C','D'], 'Value1':[1,2,3,4]})
df2 = pd.DataFrame({'key':[ 'A','B','C','E'], 'Value1':[4,5,6,7]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='key',how='inner')  ##inner join used here
```

```python
pd.merge(df1,df2,on='key',how='outer')
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,on='key',how='right')
```
